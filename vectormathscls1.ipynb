```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,7, scale_units='xy', angles ='xy',scale =0.9)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,7, scale_units='xy', angles ='xy',scale =0.9 , color ='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,7, scale_units='xy', angles ='xy',scale =0.9 , color ='b')
plt.quiver(1,0,4,7, scale_units='xy', angles ='xy',scale =0.9 , color ='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])
sum = vector1 - vector2
print(sum)
```

```python
plt.quiver(1,0,2,3, scale_units='xy', angles ='xy',scale =0.9 , color ='b')
plt.quiver(1,0,3,-2, scale_units='xy', angles ='xy',scale =0.9 , color ='r')
plt.quiver(1,0,-1,5, scale_units='xy', angles ='xy',scale =0.9 , color ='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**`multipying a vector by a scalar`**

```python
vector1 = np.asarray([0,0,2,3])
scalar_value = -0.5
print(scalar_value*vector1)
```

```python
plt.quiver(-0,-0,2,3, scale_units='xy', angles ='xy',scale =0.9 , color ='r')
plt.quiver(-0,-0,-1,-1.5, scale_units='xy', angles ='xy',scale =0.9 , color ='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
```
