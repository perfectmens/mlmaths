#projuction of 'a' vector on 'v' vector

formula



```python
a = np.array([2 ,3])
v = np.array([8 , -9])

# magitude of "v" vector
magnitude_of_v = np.sqrt(sum(v**2))
projuction_of_a_on_v = (np.dot(a,v)/magnitude_of_v)*v
print(projuction_of_a_on_v)
```

**3 Dimension**



```python
a = np.array([2 ,3 , 10])
v = np.array([8 , -9 , 30])

# magitude of "v" vector
magnitude_of_v = np.sqrt(sum(v**2))
projuction_of_a_on_v = (np.dot(a,v)/magnitude_of_v)*v
print(projuction_of_a_on_v)
```
